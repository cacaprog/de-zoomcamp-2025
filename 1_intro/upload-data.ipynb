Creating the schema for the database. The _schema_ is the structure of the database; in this case it describes the columns of our table. Pandas can output the SQL _DDL_ (Data definition language) instructions necessary to create the schema.

```python
# let's provide a name for the table
# first, we'll only output the instructions to create the table
print(pd.io.sql.get_schema(df, name='yellow_taxi_data'))
```

We have the DDL instructions but we still need to specify it for Postgres to connect and create the table. We will use `sqlalchemy`. 

Config the postgresql on terminal:

`sudo -u postgres psql -c "CREATE DATABASE ny_taxi;"`

`sudo -u postgres psql -c "CREATE USER admin WITH PASSWORD 'mysecretpass';"`

`sudo -u postgres psql -c "GRANT ALL PRIVILEGES ON DATABASE ny_taxi TO admin;"`

Connect the database

```python
from sqlalchemy import create_engine
import sys
import time

def create_db_connection(retries=5):
    for attempt in range(retries):
        try:
            engine = create_engine('postgresql://admin:mysecretpass@localhost:5432/ny_taxi')
            engine.connect()
            print("Conexão estabelecida com sucesso!")
            return engine
        except Exception as e:
            print(f"Tentativa {attempt + 1} de {retries}: {str(e)}")
            if attempt < retries - 1:
                time.sleep(2)
            else:
                print("Falha ao conectar ao banco de dados")
                sys.exit(1)

engine = create_db_connection()
```

Iterator - Let's create an iterator to read the parquet data in chunks and insert it into the database

```python
# Create an iterator to read the parquet data in chunks and insert it into the database
from time import time

df_iter = pd.read_parquet('yellow_tripdata_2024-01.parquet', engine='pyarrow')

# definy chunk size
chunk_size = 100000

# iterate over the chunks
while len(df_iter) > 0:
    # start timer
    t_start = time()
    
    # insert chunk into the database
    df_iter.head(chunk_size).to_sql(name='yellow_taxi_data', 
                                   con=engine, 
                                   if_exists='append',
                                   index=False)
    
    # remove chunk from
    df_iter = df_iter.iloc[chunk_size:]
    
    # stop timer
    t_end = time()
    
    print('Chunk inserted, took: %.3f seconds' % (t_end - t_start))
```

You can now use `pgcli -h localhost -p 5432 -u root -d ny_taxi` on a separate terminal to look at the database:

- \dt for looking at available tables.
- \d yellow_taxi_data for describing the new table.
